```python
q1=nigeria_real_gdp[:7]
q2=nigeria_real_gdp[7:14]
q3=nigeria_real_gdp[14:21]
q4=nigeria_real_gdp[21:]
```

```python
ig,ax=plt.subplots(figsize=(40,20))

plt.subplot(2,2,1)
sns.barplot(x='Year',y='Quarter_value',data=q1,color='blue',label='Quarter 1')
plt.legend(loc=3)

plt.subplot(2,2,2)
sns.barplot(x='Year',y='Quarter_value',data=q2,color='red',label='Quarter 2')
plt.legend(loc=3)

plt.subplot(2,2,3)
sns.barplot(x='Year',y='Quarter_value',data=q3,color='green',label='Quarter 3')
plt.legend(loc=3)

plt.subplot(2,2,4)
sns.barplot(x='Year',y='Quarter_value',data=q4,color='black',label='Quarter 4')
plt.legend(loc=3)
plt.savefig('quarter')
```
